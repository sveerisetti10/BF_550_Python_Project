```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import skbio 
import seaborn as sns
sns.set()
import scipy.stats as stats
# PCA imports
from sklearn.decomposition import PCA
```

```python
#We can read in our dataframe using Pandas. 
data = pd.read_csv('/Users/sveerisetti/Desktop/Boston_University/Semester2/BF_550_Python/Final_Project/Bee_data/CSV_PCoA_Data/Weighted_PCOA_Data.csv')
data


#We want to make our rownames equivalent to the Gene Name so that we can pass the dataframe through the PCA value function.
data2 = data.set_index("Gene")
data2
```

```python
#Next we want to find out the Principal Component Analysis
pcoa_values = skbio.stats.ordination.pcoa(data2.values)
pc_plot_coordinates = pcoa_values.samples[['PC1', 'PC2']]

pc_plot_coordinates
```

```python
# Here we create a list of our conditions so that we can group our data when we plot 
conditions = [
    (data['Gene'].str.contains("HC") & data['Gene'].str.contains("D3")),
    (data['Gene'].str.contains("HC") & data['Gene'].str.contains("D5")),
    (data['Gene'].str.contains("HC") & data['Gene'].str.contains("D7")),
    (data['Gene'].str.contains("HT") & data['Gene'].str.contains('D3')), 
    (data['Gene'].str.contains("HT") & data['Gene'].str.contains('D5')),
    (data['Gene'].str.contains("HT") & data['Gene'].str.contains('D7')),
    (data['Gene'].str.contains("HT") & data['Gene'].str.contains('H0')),
    (data['Gene'].str.contains("HC") & data['Gene'].str.contains('H0'))
    
    ]

# Now we can create a list of the values we want to assign for each condition
values = ["Control- Post Day 3", "Control- Post Day 5", "Control- Post Day 7", "Treatment- Post Day 3", "Treatment- Post Day 5", "Treatment- Post Day 7",
         'Treatment- Post Day 0', 'Control- Post Day 0']

# Here we create a new column and use np.select to assign values to it using our lists as arguments
data['Group'] = np.select(conditions, values)

# Now we need to extract the Group column and join it with the data df, which has all of the PCA coordinates 
extracted_col = data["Group"]
extracted_col

pc_plot_coordinates['Group'] = data['Group'].values
pc_plot_coordinates
```

```python
# Here we create a list of our conditions so that we can group our data when we plot 
conditions2 = [
    (data['Gene'].str.contains("HC") | data['Gene'].str.contains('P')),
    (data['Gene'].str.contains("HT") )  
    ]

# Here we create a list of the values we want to assign for each condition
values2 = ['Control', "Treated"]

# create a new column and use np.select to assign values to it using our lists as arguments
data['Variable'] = np.select(conditions2, values2)

#Now we need to extract the Group column and join it with the data df, which has all of the PCA coordinates 
extracted_col2 = data["Variable"]
extracted_col2

pc_plot_coordinates['Variable'] = data['Variable'].values
pc_plot_coordinates

#pc_plot_coordinates.to_csv('/Users/sveerisetti/Desktop/mom.csv')
```

```python
#Invert x values (PC1) to match that of the paper
pc_plot_coordinates["PC1"]= -1 * pc_plot_coordinates["PC1"]
```

```python
#Here we want to use seaborn in order to create our plot. We can use the markers function in order to assign each group to a point shape. 
#This is also plotted in R Studio. In R Studio we can see the ellipses. 
figure = sns.lmplot('PC1', 'PC2', data = pc_plot_coordinates, hue = 'Group', markers = ['^', '>', '<', 'p', 'v', 'o', 'h', 'H', '*'], ci = None, fit_reg = False)
figure.set(title='PC1 Vs. PC2')
figure.set(xlim=(-0.5, 0.5))
figure.set(ylim=(-0.3, 0.3))
```
