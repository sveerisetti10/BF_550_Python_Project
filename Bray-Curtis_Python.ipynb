```python
import pandas as pd
import skbio 
import seaborn as sns
sns.set()
import scipy.stats as stats
from skbio.stats import subsample_counts
from scipy.spatial.distance import braycurtis
from scipy.spatial import distance_matrix
from skbio.diversity import beta_diversity
```

```python
#Here we want to load and filter our dataframe based on the group, which in this case is Control0 
control_0 = pd.read_csv('/Users/sveerisetti/Desktop/trial.csv').dropna()
control_0 = control_0[control_0['Group'].isin(["Control0"])]
samples = control_0.Sample_ID
del control_0["Sample_ID"]
del control_0["Group"]
del control_0["Sample_ID.1"]
del control_0["Variable"]
del control_0["Shannon_Index"]
```

```python
def absolute_value(x):
    "https://en.wikipedia.org/wiki/Diversity_index"
    # Get proportions (relative abundance)
    p = abs(max(x)-min(x))
    return p
```

```python
#Here we can create our distance matrix from our relative abundance matrix
control_0 = pd.DataFrame(distance_matrix(control_0.values, control_0.values), index = samples, columns = samples)

#Here we can calculate the Bray-Curtis Dissimilarity
control_0['Index'] = sum(control_0.apply(absolute_value, axis=1)) / control_0.sum(axis=1)
control_0['Group'] = 'Control0'
control_0['Variable'] = 'C'

control_0_new = control_0[['Index', 'Group', 'Variable']]
control_0_new
#control0_distance$Index <- sum(apply(control0_distance, 1, function(x) abs(max(x)-min(x)))) / rowSums(control0_distance)
```
